```python
import pandas as pd
import os
import math
from tqdm import tqdm
import numpy as np
```

```python
from Bio import Entrez
import xml.etree.ElementTree as ET

def pmid2pmcid(email, pmid):
    Entrez.email = email

    handle = Entrez.elink(dbfrom="pubmed", db="pmc", linkname="pubmed_pmc", id=pmid, retmode="text")

    handle_read = handle.read()
    handle.close()

    root = ET.fromstring(handle_read)

    pmcid = ""

    for link in root.iter('Link'):
        for id in link.iter('Id'):
            pmcid = id.text
    return pmcid 
```

```python
df_espb = pd.read_csv('ESPB.csv')
```

```python
def get_pmcid(doi):
    cmd = """curl 'https://eutils.ncbi.nlm.nih.gov/entrez/eutils/esearch.fcgi?db=pubmed&term={}&format=json' | jq '.esearchresult.idlist[]' | tr -d '"'""".format(doi)
    id = os.popen(cmd).read().strip()
    return id
```

```python
c = 0
pmcids_list = []
for i, row in tqdm(df_espb.iterrows()):
    doi = row.DOI
    if isinstance(row.DOI, float):
        c += 1
        pmcids_list.append(np.nan)
    else:
        try:
            pmcid = get_pmcid(doi) 
            #print(pmcid)
            pmcids_list.append(int(pmcid))
            print(pmcids_list)
        except Exception as e:
            #print(e)
            pmcids_list.append(np.nan)
            c += 1
print(c)
```

```python
df_espb['PMID'] = pmcids_list
```

```python
c = 0
for x in pmcids_list:
    if np.isnan(x):
        c += 1
print(c)
```

```python
pmcids = []
for pmid in tqdm(pmcids_list):
    pmcid = pmid2pmcid('**@stanford.edu', str(pmid))
    if len(pmcid) == 0:
        pmcids.append(np.nan)
    else:    
        pmcids.append(int(pmcid))
```

```python
c = 0
for x in pmcids:
    if np.isnan(x):
        c += 1
print(c)
```

```python
df_espb['PMC ID'] = pmcids
```

```python
df_espb.to_csv('ESPB_imputed.csv', index=False)
```
